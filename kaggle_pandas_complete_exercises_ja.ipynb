````python
# 章のセットアップ: このNotebook内だけで練習用データを作成します。
# この章で使う練習用ワインレビューデータをNotebook内で作成します。
# 外部ファイル、Kaggle API、Google Driveは使用しません。
countries = ["Italy", "US", "France", "Spain", "Australia", "New Zealand", "Canada", "Germany"]
provinces = ["Tuscany", "California", "Bordeaux", "Rioja", "Victoria", "Marlborough", "Ontario", "Mosel"]
varieties = ["Sangiovese", "Pinot Noir", "Chardonnay", "Tempranillo", "Shiraz", "Sauvignon Blanc", "Riesling", "Merlot"]
tasters = ["Anna Lee", "Ben Stone", "Carla Ruiz", "Daisuke Mori"]
handles = ["@anna", "@ben", "@carla", "@daisuke"]

reviews = pd.DataFrame({
    "country": [countries[i % len(countries)] for i in range(130)],
    "description": [
        ("A tropical and fruity wine" if i % 6 == 0 else
         "A tropical wine with citrus notes" if i % 3 == 0 else
         "A fruity wine with a dry finish" if i % 2 == 0 else
         "A structured wine with mineral notes")
        for i in range(130)
    ],
    "designation": [f"Reserve {i % 5 + 1}" for i in range(130)],
    "points": [80 + (i % 21) for i in range(130)],
    "price": [np.nan if i % 17 == 0 else float(5 + (i % 40) * 3) for i in range(130)],
    "province": [provinces[i % len(provinces)] for i in range(130)],
    "region_1": [None if i % 11 == 0 else f"Region {i % 12 + 1}" for i in range(130)],
    "region_2": [None if i % 4 == 0 else f"Locale {i % 6 + 1}" for i in range(130)],
    "variety": [varieties[i % len(varieties)] for i in range(130)],
    "winery": [f"Winery {i % 15 + 1}" for i in range(130)],
    "title": [f"Sample Wine {i:03d}" for i in range(130)],
    "taster_name": [tasters[i % len(tasters)] for i in range(130)],
    "taster_twitter_handle": [handles[i % len(handles)] for i in range(130)],
})

reviews.head()
````
